### Обработка категориальных признаков

```python
# Создаем список категориальных признаков к замене их значений на числовые
category_cols = ['gender', 'geography']
```

```python
# Создаем функцию, которая на вход принимает имя выборки и спиок категориальных признаков, в которых кодирует категории в числа
# кодировщиком OneHotEncoder и заменяет изменяемый признак новыми колонками
def ohe_encoder (sample, list_cols):
    encoder = OneHotEncoder(handle_unknown='error', sparse = False, drop='first') # создание объекта кодировщика OneHotEncoder
    encoder_fitted = encoder.fit(features_trains[list_cols])  # обучение на тренировочной выборке

    categories = [] # создание списка для заполнения его названиями категорий 
    #print(encoder_fitted.categories_)
    for ar in encoder_fitted.categories_:
        ar = list(ar)
        del ar[0]  # удаление первого признака, так как используем в OneHotEncoder drop='first'
        categories += list(ar) # заполнение списка названиями категорий всех категориальных признаков
    
    res = pd.DataFrame(encoder_fitted.transform(sample[list_cols]), columns=categories) #  преобразование категории в числовые значения   
    final_df = sample.join(res) # добавление в выборку новые стобцы категорий с измененными значениями
   
    for col in list_cols:
        del final_df[col] # удаление категориальных признаков, которые были преобразованы
    return final_df
```

```python
# Преобразуем категориальные переменные в выборках кодировщиком OneHotEncoder, применяя функцию ohe_encoder
for key in samples_dict:
    samples_dict[key] = ohe_encoder(samples_dict[key], category_cols)
```

### Обучим модели без учёта дисбаланса 

```python

# Функция для получения лучших параметров (из заданных для них диапазонов) указанной модели
# На вход подется модель машинного обучения и словарь параметров с заданным диапазоном
# На выходе получаем модель с лучшими параметрами, обученную на выборке train, 
# и метрику f1

def best_model (model, param_dict, features, targets): # Подбираем гиперпараметры модели с помощью GridSearchCV
    
    model_best = GridSearchCV(estimator=model, param_grid=param_dict, cv=11, n_jobs=-1, verbose=1, scoring='f1')# refit='f1') #make_scorer(f1_score))
                              
    model_best.fit(features, targets)  # Обучение модели на заданных выборках
    
    #predict_test_gs = model_best.best_estimator_.predict(samples_dict['features_test']) # предсказание модели на тестовой выборке
    #f1_test = f1_score(target_test, predict_test_gs)  # F1-мера на тестовой выборке

    # График ROC-кривой случайной модели и модели с лучшими параметрами на тестовой выборке
    probabilities_tv = model_best.best_estimator_.predict_proba(samples_dict['features_trains']) 
    probabilities_one_tv = probabilities_tv[:, 1] # вероятности положительного класса
    auc_roc = roc_auc_score(target_trains, probabilities_one_tv)  # auc_roc - площадь под ROC-кривой   
    fpr, tpr, thresholds = roc_curve(target_trains, probabilities_one_tv)
    plt.figure()
    plt.plot(fpr, tpr, color="darkorange") 
    plt.plot([0, 1], [0, 1], linestyle='--') # ROC-кривая случайной модели
    plt.xlim([-0.02, 1.0])
    plt.ylim([0.0, 1.02])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая") 
    plt.grid()
    plt.show()    
    
    model = model_best.best_estimator_ # параметры лучшей модели, подобранные на выборке 80% от данных - на выборке train
    f1_train = model_best.best_score_ # f1_score на выборке train
        
    print ('Лучшая модель:', model, \
           '\n Лучшие параметры модели на выборке train:', model_best.best_params_, \
           '\n F1-мера на выборке train:', f1_train, \
           '\n auc_roc на выборке train:', round(auc_roc,3), \
#           '\n F1-мера на выборке test:', round(f1_test,3), '\n'
           
           ) 
      
    return model, model_best.best_params_, f1_train #, f1_test
```

### 3.2 Обучим модели после upsampling -  увеличение выборки с меньшим классом целевого признака

```python
# Функция ребалансировки классов
# на входе 3 параметра
# features - заданные выборка признаков без целевого
# target - выборка целевого признака
# direction - направление ребалансировки выборки
# на выходе выборки с отбалансированными классами

def rebalance(features, target, direction):
    
    # определяем доминирующий класс и инвертируем направление, если нужно
    if len(features[target == 0]) < len(features[target == 1]):
        direction = direction ^ 1
        
    val0 = direction ^ 1
    val1 = direction
    
    # разделяем выборки на 2 класса
    features_zeros = features[target == val0]
    features_ones = features[target == val1]
    target_zeros = target[target == val0]
    target_ones = target[target == val1]
    print(f'Количество объектов класса {val0} и {val1} соответственно до балансировки:', \
          len(features[target == val0]), len(features[target == val1]))
    
    # подсчитываем коэффициент диcбаланса между классами для получения коэффициента балансировки
    frepeat = (len(target_zeros)/len(target_ones))
    repeat = int(frepeat) # целая часть коэффициента ребалансировки
    frepeat = frepeat - repeat
    addidx = int(frepeat * len(features_ones)) #  число строк к добавлению к добалансируемому классу, т.к. repeat нецелое число
    #print(repeat)
    
    # изменяем баланс класса в указанную сторону
    features_resampled = pd.concat([features_zeros] + [features_ones] * repeat + [features_ones[0:addidx]])
    target_resampled = pd.concat([target_zeros] + [target_ones] * repeat + [target_ones[0:addidx]])
    
    # проверка баланса классов после балансировки
    print(f'Количество объектов класса {val0} и {val1} соответственно после балансировки:', \
          len(target_resampled[target_resampled == val0]), len(target_resampled[target_resampled == val1])
         )
    
    # перемешиваем выборки
    features_resampled, target_resampled = shuffle(features_resampled, target_resampled, random_state=12345) 
    
    print(f'\n Размер отбалансированных выборок: \n {features_resampled.shape} \n {target_resampled.shape}')
       
    return features_resampled, target_resampled
```
